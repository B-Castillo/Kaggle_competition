```python
import pickle
import pandas as pd
```

```python
df = pd.read_csv("../data/test.csv")
df.head(2)
```

```python
df = df[['id', 'carat', 'cut', 'color', 'clarity', 'table', 'x', 'y', 'z']]
```

```python
# encoding Clarity
with open('../data/modelo_2/encoding_clarity.pkl', 'rb') as clar:
    clar = pickle.load(clar)
    
    
# encoding Clarity
with open('../data/modelo_2/encoding_color.pkl', 'rb') as color:
    color = pickle.load(color)
    

# encoding Clarity
with open('../data/modelo_2/encoding_cut.pkl', 'rb') as cut:
    cut = pickle.load(cut)
    

# estandarización
with open('../data/modelo_2/estandarizacion.pkl', 'rb') as estan:
    estandar = pickle.load(estan)
```

```python
with open('../data/modelo_2/mejor_modelo.pkl', 'rb') as modelo:
    modelo = pickle.load(modelo)
```

```python
estandarizadas = pd.DataFrame(estandar.transform(df[['carat', 'table', 'x', 'y', 'z']]), columns = ['carat', 'table', 'x', 'y', 'z'])
estandarizadas
```

```python
clarity = pd.DataFrame(clar.transform(df[["clarity"]]), columns= ["clarity_mapeada"])
color_ = pd.DataFrame(color.transform(df[["color"]]), columns= ["color_mapeada"])
cut_ = pd.DataFrame(cut.transform(df[["cut"]]), columns= ["cut_mapeada"])
```

```python
df_final = pd.concat([estandarizadas, clarity, color_, cut_], axis = 1)
df_final.head(3)
```

```python
df_final = df_final.reindex(['carat', 'table', 'x', 'y', 'z', 'cut_mapeada', 'color_mapeada', 'clarity_mapeada'], axis=1)
df_final.head(3)
```

```python
resultados = modelo.predict(df_final)
```

```python
ids = range(0, 13485)
```

```python
final = {"id": ids,
         "price": resultados}
```

```python
df_res = pd.DataFrame(final)
```
